```python
def get_browser_page(url):

    from selenium import webdriver
    from selenium.webdriver.chrome.options import Options
    import time
    
    chrome_options = Options()
    chrome_options.add_argument("--start-maximized");

    browser = webdriver.Chrome("/usr/local/bin/chromedriver",options = chrome_options) 
    browser.get(url)

    browser.find_element_by_xpath('//input[@value = "Search"]').click()
    time.sleep(5)

    try:
        browser.find_element_by_partial_link_text('Show all').click()
        time.sleep(5)
    except:
        print("No show all button")

        
    page     =  browser . page_source
    permits  =  pd . read_html ( page )
    
    browser . close ( )
    
    return permits
        
    
def extract_ArlingtonMA_permits(year):
    import numpy as np
    #from datetime import datetime

    url = 'http://arlserver.town.arlington.ma.us/BuildingPermits/Select.pl?permittype=&issue=1/1/{year}&issue_thru=12/31/{year}'.format(year=year)
    
    permits = get_browser_page(url)

    permits = permits[1]   
    permits.columns = ['streetNum'] + list(permits.columns[1:])

    
    permits[['streetNum','Address']] = permits[['streetNum','Address']].ffill().replace(np.nan,'', regex=True)
    #permits['asof']  =  datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    
    return permits


def permits_get_solar_norm(permits):

    mask = ~pd.isnull(permits.Description) & permits.Description.str.contains('Solar|solar')

    permits = permits[mask].copy()
    
    permits.Address=permits.Address.str.upper()
    
    for col in ['streetNum','Owner']:
        permits[col]=permits[col].astype(str)
    
    permits = permits[permits.Issued!='2001']  #85 	Oakland Ave 	2001 	B 	R 	23 	remodel kitchen 	Trinity Baptist Church 	David 	7000.0 	105.0 
    permits['Issued']=pd.to_datetime(permits['Issued'],format='%m-%d-%Y').dt.strftime('%Y-%m-%d')

    permits=permits.sort_values('Issued').reset_index(drop=True)

    if permits.Value.dtype==type(str):
        permits.loc[permits.Value.str.contains(' '),'Value']='0'
        permits.Value=permits.Value.astype(float)

    if permits.Fee.dtype==type(str):
        permits.loc[permits.Fee.str.contains(' '),'Fee']='0'
        permits.Fee=permits.Fee.astype(float)

    #permits['installer'] = permits['Contractor'].str.lower().str.replace(' |;|,','', regex=True)
    permits['Contractor'] = permits['Contractor'].astype(str).str.replace('nan','').fillna('')
    permits['streetNum']  = permits['streetNum'].str.replace('\#','-', regex=True)

    
    permits = permits_norm(permits)

#     summary = permits.groupby("installer").agg(
#         {"installer":len,
#          "Value":sum,
#          "Fee":sum,
#         }
#     )
#     summary.columns = ['installs','value','fee']
    
    aggregates = permits.groupby(['streetNum','Address']).agg(
    {
        "Permit":list,
        "Description":list,
        "Contractor":list,
        "Owner":max,
        "Value":sum,
        "Fee":sum,
        "Issued":[list,min]
    }).reset_index()

    aggregates.columns = ['streetNum','address','permits','descriptions','contractor','owner','permit_value','permit_fee','issued','effective_date']
    

    return aggregates#, summary


def permits_norm(permits) :

    permits.Address=permits.Address . str . replace ( 'PARK$','PK', regex=True) \
                    . str . replace ( ' WAY$',' WY', regex=True) \
                    . str . replace ( '^NEWMAN WY$','NEWMAN WAY', regex=True) \
                    . str . replace ( '^BRATTLE TER$','BRATTLE TERR', regex=True) \
                    . str . replace ( '^APACHE TRL$','APACHE TR', regex=True) \
                    . str . replace ( '^FARMER CIR$|^FARMER\'S CIR$','FARMER`S CIR', regex=True) \
                    . str . replace ( '^GRANDVIEW RD$','GRAND VIEW RD', regex=True) \
                    . str . replace ( '^LOMBARD TER$','LOMBARD TERR', regex=True) \
                    . str . replace ( '^MOUNT VERNON ST$','MT. VERNON ST', regex=True) \
                    . str . replace ( '^N UNION ST$','NORTH UNION ST', regex=True) \
                    . str . replace ( '^PARK AVENUE EXT$','PARK AVE EXT', regex=True) \
                    . str . replace ( '^UPLAND RD$|^UPLAND RD W$','UPLAND RD WEST', regex=True) \
                    . str . replace ( '^WALNUT TER$','WALNUT TERR', regex=True) \
                    . str . replace ( '^WYMAN TER$','WYMAN TERR', regex=True) \
                    . str . replace ( '^SUMMER STREET PL$','SUMMER ST PL', regex=True) \
                    . str . replace ( '^RUSSELL TER$','RUSSELL TERR', regex=True) \
                    . str . replace ( '^LORRAINE TER$','LORRAINE TERR', regex=True) \
                    . str . replace ( '^LOWELL STREET PL$','LOWELL ST PL', regex=True) \
                    . str . replace ( '^PARK STREET PL$','PARK ST PL', regex=True) \
                    . str . replace ( '^WEST COURT TE$','WEST COURT TERR', regex=True) \
                    . str . replace ( '^LEE TER$','LEE TERR', regex=True) #\
#                     . str . replace ( '^$','', regex=True) \
#                     . str . replace ( '^$','', regex=True) \
    
    mask = (permits.Address == 'WASHINGTON ST') & (permits.streetNum=='73')
    permits.loc[mask,'streetNum'] = '71'
    mask = (permits.Address == 'CENTRAL ST') & (permits.streetNum=='10')
    permits.loc[mask,'streetNum'] = '8'
    mask = (permits.Address == 'ADAMS ST') & (permits.streetNum=='56-1')
    permits.loc[mask,'streetNum'] = '56'
    mask = (permits.Address == 'HERBERT RD') & (permits.streetNum=='84')
    permits.loc[mask,'streetNum'] = '82'
    mask = (permits.Address == 'BOWDOIN ST') & (permits.streetNum=='13')
    permits.loc[mask,'streetNum'] = '11'
    mask = (permits.Address == 'ALBERMARLE ST') & (permits.streetNum=='16')
    permits.loc[mask,'streetNum'] = '14'
    mask = (permits.Address == 'APPLETON PL') & (permits.streetNum=='377')
    permits.loc[mask,'Address'] = 'APPLETON ST'
    mask = (permits.Address == 'UPLAND RD WEST') & ((permits.streetNum=='11')|(permits.streetNum=='39'))
    permits.loc[mask,'Address'] = 'UPLAND RD'
    mask = ((permits.Address == 'UPLAND RD')|(permits.Address == 'UPLAND RD W')) & ((permits.streetNum=='45'))
    permits.loc[mask,'Address'] = 'UPLAND RD WEST'
    mask = (permits.Address == 'MASS AVE') & ((permits.streetNum=='626'))
    permits.loc[mask,'streetNum'] = '616'
    mask = (permits.Address == 'MASS AVE') & ((permits.streetNum=='869'))
    permits.loc[mask,'streetNum'] = '855'
    mask = (permits.Address == 'BROADWAY') & ((permits.streetNum=='37'))
    permits.loc[mask,'streetNum'] = '33'
    mask = (permits.Address == 'BROADWAY') & ((permits.streetNum=='297'))
    permits.loc[mask,'streetNum'] = '295'
    mask = (permits.Address == 'BROADWAY') & ((permits.streetNum=='144'))
    permits.loc[mask,'streetNum'] = '142'
    mask = (permits.Address == 'PARK AVE EXT') & ((permits.streetNum=='83')|(permits.streetNum=='82')|(permits.streetNum=='85'))
    permits.loc[mask,'streetNum'] = '99'
    mask = (permits.Address == 'NORTH UNION ST') & ((permits.streetNum=='60'))
    permits.loc[mask,'Address'] = 'EVERETT ST'
    permits.loc[mask,'streetNum'] = '187'
    mask = (permits.Address == 'NORTH UNION ST') & ((permits.streetNum=='95'))
    permits.loc[mask,'streetNum'] = '91'
    mask = (permits.Address == 'EGERTON RD') & ((permits.streetNum=='93'))
    permits.loc[mask,'streetNum'] = '91'
    mask = (permits.Address == 'MILTON ST') & ((permits.streetNum=='90-1'))
    permits.loc[mask,'streetNum'] = '90'
    mask = (permits.Address == 'BURCH ST') & ((permits.streetNum=='87'))
    permits.loc[mask,'Address'] = 'BEVERLY RD'
    mask = (permits.Address == 'PARK HILL CIR') & ((permits.streetNum=='10'))
    permits.loc[mask,'Address'] = 'SUMMER HILL CIR'
    mask = (permits.Address == 'PALMER ST') & ((permits.streetNum=='10-2'))
    permits.loc[mask,'streetNum'] = '10'
    mask = (permits.Address == 'NEWPORT ST') & ((permits.streetNum=='85-2'))
    permits.loc[mask,'streetNum'] = '83'
    mask = (permits.Address == 'MILTON ST') & ((permits.streetNum=='105-107'))
    permits.loc[mask,'streetNum'] = '105'
    mask = (permits.Address == 'PALMER ST') & ((permits.streetNum=='102'))
    permits.loc[mask,'streetNum'] = '100'
    mask = (permits.Address == 'AMHERST ST') & ((permits.streetNum=='10'))
    permits.loc[mask,'streetNum'] = '8'
    mask = (permits.Address == 'BROOKS AVE') & ((permits.streetNum=='80'))
    permits.loc[mask,'streetNum'] = '78'
    mask = (permits.Address == 'BROOKS AVE') & ((permits.streetNum=='84'))
    permits.loc[mask,'streetNum'] = '82'
    mask = (permits.Address == 'DOROTHY RD') & ((permits.streetNum=='85'))
    permits.loc[mask,'streetNum'] = '83'
    mask = (permits.Address == 'BOW ST') & ((permits.streetNum=='82'))
    permits.loc[mask,'streetNum'] = '84'
    mask = (permits.Address == 'EGERTON RD') & ((permits.streetNum=='8'))   ##location==8
    permits.loc[mask,'streetNum'] = '6'
    mask = (permits.Address == 'FREEMAN ST') & ((permits.streetNum=='10-2'))   
    permits.loc[mask,'streetNum'] = '10'
    mask = (permits.Address == 'GRAFTON ST') & ((permits.streetNum=='107'))   
    permits.loc[mask,'streetNum'] = '105'
    mask = (permits.Address == 'TEMPLE ST') & ((permits.streetNum=='11-1'))   ##SITE_ADDR = 11 A TEMPLE ST
    permits.loc[mask,'streetNum'] = '11'
    mask = (permits.Address == 'MARRIGAN ST') & ((permits.streetNum=='11.0'))   
    permits.loc[mask,'streetNum'] = '11'
    mask = (permits.Address == 'OAKLAND AVE') & ((permits.streetNum=='108'))  
    permits.loc[mask,'Address'] = 'CEDAR AVE'
    permits.loc[mask,'streetNum'] = '6'
    mask = (permits.Address == 'PARK ST') & ((permits.streetNum=='78'))   ##location==78
    permits.loc[mask,'streetNum'] = '76'
    mask = (permits.Address == 'AMSDEN ST') & ((permits.streetNum=='73'))
    permits.loc[mask,'streetNum'] = '71'
    mask = (permits.Address == 'OSBORNE RD') & ((permits.streetNum=='7'))
    permits.loc[mask,'streetNum'] = '5'
    mask = (permits.Address == 'MILTON ST') & ((permits.streetNum=='69'))   ##LOCATION==1
    permits.loc[mask,'streetNum'] = '67'
    mask = (permits.Address == 'BROOKS AVE') & ((permits.streetNum=='68'))   
    permits.loc[mask,'streetNum'] = '66'
    mask = (permits.Address == 'PALMER ST') & ((permits.streetNum=='110'))   
    permits.loc[mask,'streetNum'] = '108'
    mask = (permits.Address == 'THORNDIKE ST') & ((permits.streetNum=='130-1'))   
    permits.loc[mask,'streetNum'] = '130'
    mask = (permits.Address == 'MOTT ST') & ((permits.streetNum=='62-1')) 
    permits.loc[mask,'streetNum'] = '62'
    mask = (permits.Address == 'FRANKLIN ST') & ((permits.streetNum=='141')) 
    permits.loc[mask,'streetNum'] = '139'
    mask = (permits.Address == 'FRANKLIN ST') & ((permits.streetNum=='141-2')) 
    permits.loc[mask,'streetNum'] = '139'
    mask = (permits.Address == 'TWIN CIRCLE DR') & ((permits.streetNum=='6.0')) 
    permits.loc[mask,'streetNum'] = '6'
    mask = (permits.Address == 'MARY ST') & ((permits.streetNum=='143.0')) 
    permits.loc[mask,'streetNum'] = '143'
    mask = (permits.Address == 'EVERETT ST') & ((permits.streetNum=='63-1')) 
    permits.loc[mask,'streetNum'] = '63'
    mask = (permits.Address == 'WALNUT CT') & ((permits.streetNum=='62'))
    permits.loc[mask,'Address'] = 'WALNUT ST'
    mask = (permits.Address == 'DOROTHY RD') & ((permits.streetNum=='63'))   ##LOCATION=63
    permits.loc[mask,'streetNum'] = '61'
    mask = (permits.Address == 'HIBBERT ST') & ((permits.streetNum=='57.0')) 
    permits.loc[mask,'streetNum'] = '57'
    mask = (permits.Address == 'MAGNOLIA ST') & ((permits.streetNum=='57-2')) 
    permits.loc[mask,'streetNum'] = '55'
    mask = (permits.Address == 'CYPRESS RD') & ((permits.streetNum=='54-1')) 
    permits.loc[mask,'streetNum'] = '54'
    mask = (permits.Address == 'FOREST ST') & ((permits.streetNum=='52')) 
    permits.loc[mask,'streetNum'] = '50'
    mask = (permits.Address == 'MILTON ST') & ((permits.streetNum=='15')) 
    permits.loc[mask,'streetNum'] = '13'
    mask = (permits.Address == 'GROVE ST') & ((permits.streetNum=='15')) 
    permits.loc[mask,'streetNum'] = '13'
    mask = (permits.Address == 'HERBERT RD') & ((permits.streetNum=='148')) ##location==2
    permits.loc[mask,'streetNum'] = '146'
    mask = (permits.Address == 'OAK KNL') & ((permits.streetNum=='15')) 
    permits.loc[mask,'Address'] = 'OAK KNOLL'
    mask = (permits.Address == 'MELROSE ST') & ((permits.streetNum=='54-1')) 
    permits.loc[mask,'Address'] = 'CROSBY ST'
    permits.loc[mask,'streetNum'] = '54'
    mask = (permits.Address == 'PARK AVE') & ((permits.streetNum=='46')) 
    permits.loc[mask,'ADDRESS'] = 'PARK AVE EXT'
    permits.loc[mask,'streetNum'] = '46'
    mask = (permits.Address == 'PARK AVE EXT') & ((permits.streetNum=='48')) 
    permits.loc[mask,'streetNum'] = '46'
    mask = (permits.Address == 'WALDO RD') & ((permits.streetNum=='48')) ##location==2
    permits.loc[mask,'streetNum'] = '46'
    mask = (permits.Address == 'PHILEMON ST') & ((permits.streetNum=='15.0')) ##location==2
    permits.loc[mask,'streetNum'] = '15'
    mask = (permits.Address == 'MARATHON ST') & ((permits.streetNum=='47.0')) ##location==2
    permits.loc[mask,'streetNum'] = '47'
    mask = (permits.Address == 'WINDSOR ST') & ((permits.streetNum=='15-2')) ##location==2
    permits.loc[mask,'streetNum'] = '15'
    mask = (permits.Address == 'THORNDIKE ST') & ((permits.streetNum=='16')) 
    permits.loc[mask,'streetNum'] = '14'
    mask = (permits.Address == 'SHERBORN ST') & ((permits.streetNum=='46')) 
    permits.loc[mask,'streetNum'] = '44'
    mask = (permits.Address == 'MYSTIC VALLEY PKWY') & ((permits.streetNum=='44')) 
    permits.loc[mask,'streetNum'] = '42'
    mask = (permits.Address == 'SUMMER ST') & ((permits.streetNum=='150-2')) 
    permits.loc[mask,'streetNum'] = '150'
    mask = (permits.Address == 'HILLSDALE RD') & ((permits.streetNum=='157')) 
    permits.loc[mask,'Address'] = 'HILLSIDE AVE'
    mask = (permits.Address == 'EDITH ST') & ((permits.streetNum=='4')|(permits.streetNum=='4-2')) 
    permits.loc[mask,'streetNum'] = '2'
    mask = (permits.Address == 'WASHINGTON ST') & ((permits.streetNum=='38'))    ##location=38
    permits.loc[mask,'streetNum'] = '36'
    mask = (permits.Address == 'MOTT ST') & ((permits.streetNum=='36'))    ##location=36
    permits.loc[mask,'streetNum'] = '34'
    mask = (permits.Address == 'FRANKLIN ST') & ((permits.streetNum=='42-2')) 
    permits.loc[mask,'streetNum'] = '42'
    mask = (permits.Address == 'SHERBORN ST') & ((permits.streetNum=='46')) 
    permits.loc[mask,'streetNum'] = '44'
    mask = (permits.Address == 'COLEMAN RD') & ((permits.streetNum=='38-2')) 
    permits.loc[mask,'streetNum'] = '38'
    mask = (permits.Address == 'MOULTON RD') & ((permits.streetNum=='44')) 
    permits.loc[mask,'streetNum'] = '42'
    mask = (permits.Address == 'WELLINGTON ST') & ((permits.streetNum=='37-1')) ##location=1
    permits.loc[mask,'streetNum'] = '37'
    mask = (permits.Address == 'WELLINGTON ST') & ((permits.streetNum=='37-2')) ##location=2
    permits.loc[mask,'streetNum'] = '37'
    mask = (permits.Address == 'LOMBARD TERR') & ((permits.streetNum=='38')) 
    permits.loc[mask,'streetNum'] = '36'
    mask = (permits.Address == 'PARK ST PL') & ((permits.streetNum=='4'))   ##location=A/B?
    permits.loc[mask,'streetNum'] = '2'
    mask = (permits.Address == 'COLEMAN RD') & ((permits.streetNum=='35-2')) 
    permits.loc[mask,'streetNum'] = '33'
    mask = (permits.Address == 'COLEMAN RD') & ((permits.streetNum=='31')) 
    permits.loc[mask,'streetNum'] = '29'
    mask = (permits.Address == 'MAGNOLIA ST') & ((permits.streetNum=='35a')) 
    permits.loc[mask,'streetNum'] = '35'
    mask = (permits.Address == 'LOCKELAND AVE') & ((permits.streetNum=='35-37')) 
    permits.loc[mask,'streetNum'] = '35'
    mask = (permits.Address == 'CYPRESS RD') & ((permits.streetNum=='5')) 
    permits.loc[mask,'streetNum'] = '1'
    mask = (permits.Address == 'LAKEHILL AVE') & ((permits.streetNum=='35')) 
    permits.loc[mask,'streetNum'] = '33'
    mask = (permits.Address == 'COLEMAN RD') & ((permits.streetNum=='35')) 
    permits.loc[mask,'streetNum'] = '33'
    mask = (permits.Address == 'DARTMOUTH ST') & ((permits.streetNum=='33')) 
    permits.loc[mask,'streetNum'] = '31'
    mask = (permits.Address == 'FAIRMONT ST') & ((permits.streetNum=='32.0')) 
    permits.loc[mask,'streetNum'] = '32'
    mask = (permits.Address == 'THORNDIKE ST') & ((permits.streetNum=='32-2')) 
    permits.loc[mask,'streetNum'] = '30'
    mask = (permits.Address == 'THORNDIKE ST') & ((permits.streetNum=='32')) 
    permits.loc[mask,'streetNum'] = '30'
    mask = (permits.Address == 'HIGHLAND AVE') & ((permits.streetNum=='31A'))  ##location 31A 
    permits.loc[mask,'streetNum'] = '31'
    mask = (permits.Address == 'FRAZER RD') & ((permits.streetNum=='30.0'))  
    permits.loc[mask,'streetNum'] = '30'
    mask = (permits.Address == 'FAIRMONT ST') & ((permits.streetNum=='32.0'))  
    permits.loc[mask,'streetNum'] = '32'
    mask = (permits.Address == 'CEDAR AVE') & ((permits.streetNum=='173.0'))  
    permits.loc[mask,'streetNum'] = '173'
    mask = (permits.Address == 'LEE TERR') & ((permits.streetNum=='25'))   
    permits.loc[mask,'streetNum'] = '23'
    mask = (permits.Address == 'LEE TERR') & ((permits.streetNum=='29'))   ##LOCATION=29
    permits.loc[mask,'streetNum'] = '27'
    mask = (permits.Address == 'COLEMAN RD') & ((permits.streetNum=='27')) 
    permits.loc[mask,'streetNum'] = '25'
    mask = (permits.Address == 'MYSTIC VALLEY PKWY') & ((permits.streetNum=='28')) 
    permits.loc[mask,'streetNum'] = '26'
    mask = (permits.Address == 'FARMER`S CIR') & ((permits.streetNum=='27')) 
    permits.loc[mask,'Address'] = 'FARMER RD'
    mask = (permits.Address == 'PONDVIEW RD') & ((permits.streetNum=='26')) 
    permits.loc[mask,'streetNum'] = '24'
    mask = (permits.Address == 'ALFRED RD') & ((permits.streetNum=='26')) 
    permits.loc[mask,'streetNum'] = '24'
    mask = (permits.Address == 'DARTMOUTH ST') & ((permits.streetNum=='25')) 
    permits.loc[mask,'streetNum'] = '23'
    mask = (permits.Address == 'CHANDLER ST') & ((permits.streetNum=='2426'))  ##????
    permits.loc[mask,'streetNum'] = '24'
    mask = (permits.Address == 'COLEMAN RD') & ((permits.streetNum=='24-2'))  
    permits.loc[mask,'streetNum'] = '24'
    mask = (permits.Address == 'COLEMAN RD') & ((permits.streetNum=='19'))  
    permits.loc[mask,'streetNum'] = '17'
    mask = (permits.Address == 'NOURSE ST') & ((permits.streetNum=='24'))   
    permits.loc[mask,'streetNum'] = '22'
    mask = (permits.Address == 'AVON PL') & ((permits.streetNum=='24'))
    permits.loc[mask,'streetNum'] = '22'
    mask = (permits.Address == 'ALTON ST') & ((permits.streetNum=='23-2')) 
    permits.loc[mask,'streetNum'] = '21'
    mask = (permits.Address == 'LOWELL ST') & ((permits.streetNum=='23')) 
    permits.loc[mask,'Address'] = 'LOWELL ST PL'
    mask = (permits.Address == 'SURRY RD') & ((permits.streetNum=='22-2')) 
    permits.loc[mask,'streetNum'] = '20'
    mask = (permits.Address == 'SURRY RD') & ((permits.streetNum=='22')) 
    permits.loc[mask,'streetNum'] = '20'
    mask = (permits.Address == 'CHANDLER ST') & ((permits.streetNum=='20-22'))
    permits.loc[mask,'streetNum'] = '20'
    mask = (permits.Address == 'ALFRED RD') & ((permits.streetNum=='21'))  
    permits.loc[mask,'streetNum'] = '19'
    mask = (permits.Address == 'WALNUT ST') & ((permits.streetNum=='2'))  
    permits.loc[mask,'Address'] = 'WALNUT TERR'
    mask = (permits.Address == 'MYSTIC ST') & ((permits.streetNum=='199'))   
    permits.loc[mask,'streetNum'] = '197'
    mask = (permits.Address == 'WYMAN TERR') & ((permits.streetNum=='19-1'))  ##LOCATION=1
    permits.loc[mask,'streetNum'] = '19'
    mask = (permits.Address == 'WYMAN TERR') & ((permits.streetNum=='19-2'))  ##LOCATION=2
    permits.loc[mask,'streetNum'] = '19'
    mask = (permits.Address == 'BELKNAP ST') & ((permits.streetNum=='19')) 
    permits.loc[mask,'streetNum'] = '17'
    mask = (permits.Address == 'WHITE ST') & ((permits.streetNum=='18-2'))   ##LOCATION=2
    permits.loc[mask,'streetNum'] = '18'
    mask = (permits.Address == 'LOCKELAND AVE') & ((permits.streetNum=='16-2'))
    permits.loc[mask,'streetNum'] = '16'
    mask = (permits.Address == 'THOMAS ST') & ((permits.streetNum=='130-2'))  
    permits.loc[mask,'Address'] = 'THORNDIKE ST'
    permits.loc[mask,'streetNum'] = '130'
    mask = (permits.Address == 'WOLLASTON AVE') & ((permits.streetNum=='19'))  ##bad permit
    permits.loc[mask,'streetNum'] = '16'
    mask = (permits.Address == 'PARK AVE') & ((permits.streetNum=='48'))  
    permits.loc[mask,'Address'] = 'PARK AVE EXT'
    permits.loc[mask,'streetNum'] = '46'
    mask = (permits.Address == 'HIGHLAND AVE') & ((permits.streetNum=='126'))  
    permits.loc[mask,'streetNum'] = '128'
    mask = (permits.Address == 'LAKE ST') & ((permits.streetNum=='225'))  
    permits.loc[mask,'streetNum'] = '255'
    
    return permits

## extract permits for multiple years
def extract_all_permits(data_dir , start_year = 2000 , end_year = 2022 ) :
    
    permits = pd . DataFrame ( )
    
    for year in range ( start_year , end_year ) :
        
        permits = permits . append ( extract_ArlingtonMA_permits ( year ) )

    solar_permits , contractor_leaders = permits_get_solar_norm ( permits )
    
    ##print(contractor_leaders.sort_values('installs',ascending=False)[0:30].to_markdown())
    
    solar_permits . to_csv ( data_dir + 'solar_permits.tsv' , sep = '\t' , index = False )
    permits . to_csv ( data_dir + 'permits.tsv' , sep = '\t' , index = False )

    return solar_permits, contractor_leaders

def get_massgis_structures_assessor_shapefiles(data_dir = '/data/L3/L3_SHP_M010_Arlington/', town_code=10):

    url          =  'http://download.massgis.digital.mass.gov/shapefiles/structures/structures_poly_{town_code}.zip'
    structures   =  gpd . read_file ( url . format ( town_code = town_code ) ) . to_crs ( "EPSG:4326" )


    ##manual file download and merge shape and database file together
    url   =  data_dir  +  'M010TaxPar_CY21_FY21.shp'
    shp   =  gpd . read_file ( url ) . to_crs ( "EPSG:4326" )

    url   =  data_dir + 'M010Assess_CY21_FY21.dbf'
    dbf   =  gpd . read_file ( url ) . drop ( 'geometry' , axis = 1 ) . copy ( )  ## messed up geometry field all None

    combo  =  shp . merge ( dbf , on = ['TOWN_ID','LOC_ID'] , how = 'outer' , indicator = True )
    combo  =  combo [ combo . _merge == 'both' ] . reset_index ( drop = True )   ##63 no matches; water, roads, etc.
    combo  =  combo . drop ( '_merge' , axis = 1 )

    return combo, structures


def geohasher ( center ) :

    import geohash

    return geohash . encode_uint64 ( np . round ( center . y , 7 ) , np . round ( center . x , 7 ) )


def merge_gis_permits ( solar_permits, assessor, structures ) :
      
    solar_systems = solar_permits . merge ( 
        assessor , 
        right_on   =  [ 'ADDR_NUM' , 'FULL_STR' ] ,
        left_on    =  [ 'streetNum' , 'address' ] ,
        how        =    'outer' ,
        indicator  =  True
    )

    solar_systems_wo_geometry = solar_systems [ solar_systems . _merge == 'left_only' ]

    solar_systems = gpd . GeoDataFrame ( solar_systems [ solar_systems . _merge == 'both' ] ) . to_crs ( "EPSG:4326" ) . reset_index ( drop = True )

    solar_systems [ 'tokenId' ] = np . int64
    
    temp_structs = structures . to_crs(epsg=3035) . copy ( )
    temp_solars  = solar_systems . to_crs(epsg=3035) . copy ( )

    for idx in range ( len ( solar_systems ) ) :

        mask = temp_structs . centroid . intersects ( temp_solars . loc [ idx , 'geometry' ] )   ##intersects Vs within finds all structures

        if len ( structures [ mask ] ) > 0 :
            
            foo = structures [ mask ] [ structures [ mask ] . AREA_SQ_FT == structures [ mask ] . AREA_SQ_FT . max ( ) ]
            
            solar_systems . loc [ idx , 'geometry' ] =  ( foo . geometry ) . values ## NB centroid returns lon / lat
            
        else :

            mask = structures . to_crs(epsg=3035) . intersects ( solar_systems . to_crs(epsg=3035) . loc [ idx , 'geometry' ] )   ##intersects Vs within finds all structures

            if len ( structures [ mask ] ) > 0 :

                print ( idx , 'centroid of structure not inside property boundary' )  ##using centroid of plot instead of structure
                print(solar_systems[solar_systems.index.isin([idx])][['streetNum','address','effective_date','STYLE']])

                foo = structures [ mask ] [ structures [ mask ] . AREA_SQ_FT == structures [ mask ] . AREA_SQ_FT . max ( ) ]

                solar_systems . loc [ idx , 'geometry' ] =  ( foo . geometry ) . values ## NB centroid returns lon / lat

            else :
            
                print ( idx , 'No structure found' )  ##using centroid of plot instead of structure
                print(solar_systems[solar_systems.index.isin([idx])][['streetNum','address','effective_date','STYLE']])
                

            
    center    =  solar_systems['geometry'] . to_crs('+proj=cea') . centroid . to_crs ( solar_systems . crs ) #centroid 

    solar_systems['coord'] = list ( 
        zip ( 
            np . round ( center . y , 7 ) , 
            np . round ( center . x , 7 ) 
        )
    ) ## NB centroid returns lon / lat  S/B Point type

    ## old google maps "api", shoudl check out new
    url = 'http://maps.google.com/maps?t=k&q=loc:{y}+{x}'
    url = 'https://www.google.com/maps/@?api=1&map_action=map&center={y}%2C{x}&zoom=21&basemap=satellite'

    for idx in range ( len ( solar_systems ) ) :
        solar_systems . loc [ idx , 'image' ] = url . format ( y = center [ idx ] . y , x = center [ idx ] . x )


    solar_systems [ 'tokenId' ] = center . apply ( geohasher )

    #solar_systems .issued          =  solar_systems . issued . dt . strftime ( '%Y-%m-%d' )
    solar_systems . effective_date  =  pd.to_datetime(solar_systems . effective_date)# . dt . strftime ( '%Y-%m-%d' )
    solar_systems . _merge          =  solar_systems . _merge . astype ( str )

    for col in [ 'permits' , 'issued' , 'descriptions' , 'contractor' , 'coord' ] :
        solar_systems [ col ] = solar_systems [ col ] . apply ( str )  ##astype?

        
    return solar_systems , solar_systems_wo_geometry

def remove_dup_tokenId(solar_systems):
## dup matches for condo conversions.  streetNum + address + LOCATION
    address_merge_exceptions = ["14 ALBERMARLE ST 14","8 AMHERST ST 10","61 DOROTHY RD 61","6 EGERTON RD 6","94 GRAFTON ST 2","79 HARLOW ST 2","146 HERBERT RD 1",
                                "31 HIGHLAND AVE 31B","27 LEE TERR 29","36 MARGARET ST 1","374 MASS AVE 2","374 MASS AVE 3","374 MASS AVE 4","5 MELROSE ST 7",
                                "91 MELROSE ST 93","38 MENOTOMY RD 2","67 MILTON ST 2","68 MILTON ST 2","88 MILTON ST 2","34 MOTT ST 36","51 NEWCOMB ST 53",
                                "83 NEWPORT ST 83","91 NORTH UNION ST 91","76 PARK ST 78","64 RIVER ST 66","20 RUSSELL ST 1","11 TEMPLE ST ","79 THORNDIKE ST 2",
                                "36 WASHINGTON ST 38","18 WHITE ST 2","19 WYMAN TERR 3"]

    mask = pd.isnull(solar_systems.LOCATION)
    solar_systems.loc[mask,'LOCATION']=""
    solar_systems['munge'] = (solar_systems.streetNum.values + ' ' + solar_systems.address.values + ' ' + solar_systems.LOCATION.values)
    mask = solar_systems['munge'].isin(address_merge_exceptions)
    solar_systems=solar_systems[~mask]


    dups = solar_systems[solar_systems.duplicated('tokenId',keep=False)] 
    mask = dups.duplicated('permits',keep=False)

    print('\n Separate systems')
    print(dups[~mask][['streetNum','address','owner','OWNER1','USE_CODE','tokenId','LOCATION','permits']].to_markdown())
    print('\n\n Shared systems')
    print(dups[mask][['streetNum','address','owner','OWNER1','USE_CODE','tokenId','LOCATION','permits']].to_markdown())

    solar_systems['shared'] = ''

    shared = dups[mask].groupby(['tokenId']).agg({"LOCATION":sum}).reset_index()
    separate = dups[~mask].groupby(['tokenId']).agg({"LOCATION":sum,"permits":sum}).reset_index()
    separate.permits = separate.permits.astype(str).str.replace('\]\[',', ')


    for idx in range(len(shared)):
        mask = solar_systems.tokenId == shared.iloc[idx].tokenId
        solar_systems.loc[mask,'shared'] = shared.loc[idx].LOCATION
        
    for idx in range(len(separate)):
        mask = solar_systems.tokenId == separate.iloc[idx].tokenId
        solar_systems.loc[mask,'shared'] = separate.iloc[idx].LOCATION
        solar_systems.loc[mask,'permits'] = separate.iloc[idx].permits.replace('\]\[',', ')    

    
    return solar_systems[~solar_systems.duplicated('tokenId')]
```

```python
data_dir = '/PATH/TO/YOUR/PERMITS/'
data_dir = '/data/code/web/truepersons/data/public/US/MA/Arlington/permits/'
## Restore

# solar_permits  =  pd . read_csv ( data_dir  +  'solar_permits.tsv' , sep = '\t' )
permits        =  pd . read_csv ( data_dir  +  'permits.tsv' , sep = '\t' )

solar_permits  = permits_get_solar_norm ( permits )
assessor, structures = get_massgis_structures_assessor_shapefiles ( )

solar_systems , solar_systems_wo_geometry = merge_gis_permits ( solar_permits , assessor , structures ) 

solar_systems = remove_dup_tokenId ( solar_systems ) . reset_index ( drop = True )
```
